```python
import requests
import random


# Generate test data according to schema
def generate_test_data(num_samples=5):
    # List of possible commodity codes
    products = [
        "Roasted Coffee", "Refined Sugar", "Cocoa powder", "Fresh cheese",
		"Carbonated beverages", "Leather shoes", "Wheat Flour", 
		"Powdered Milk", "Cocoa Beans", "Refined Salt"
    ]
    test_data = []
    for _ in range(num_samples):
        record = {
            "description": random.choice(products),
            "quantity": round(random.uniform(50, 1000), 2),
            "price": round(random.uniform(1.5, 15.0), 2)
        }
        test_data.append(record)
    return test_data


# Generate 5 test samples
df_request = generate_test_data(5)

try:
    for i in range(len(df_request)):
        response = requests.post(
            "http://20.29.159.59:8000/v1",
            json=df_request[i]
        )
        result = response.json()
        print(df_request[i], "--> ", result)
        print("-" * 120)

except Exception as ex:
    print(f"Request: The following error occurred: {ex}")
```
